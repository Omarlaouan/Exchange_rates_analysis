## DATA CLEANING

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates.head()
```

```python
# Lets isolate the time and US dallar columns : 
euro_to_dollar = exchange_rates[['Time','US_dollar']]
```

```python
euro_to_dollar.head()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

62 times the value inside [US_dollar] is the character "-", we need to remove it

```python
# removing the rows with '-' by using a boolean mask
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
# convert the values into float
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

```python
# Now that all the values are numerical, we can start doing some analyzes.
```

#### Lets see how the financial crisis affected the exchange rate between Euro and US dollar. 

```python
import matplotlib.pyplot as plt
# To enables jupyter notebook to display our graphs :
%matplotlib inline 

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

We can notive that there is a lot of variations daily. For our purpose, it is better to smooth the plot. we can use the rolling mean (also called moving average).Lets use a rolling window of 30 days

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
euro_to_dollar
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.show()
```

# Data visualization
Lets take the window of 2007-2008 to see the impact of the financial crisis on exchange rates. 



```python
financial_crisis = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2006
                   ) & (euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_7_8 = euro_to_dollar.copy(
                   )[(euro_to_dollar.Time.dt.year >= 2007
                   ) & (euro_to_dollar.Time.dt.year < 2009)]
```

```python
# Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

# Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'],
        financial_crisis['rolling_mean'],
        linewidth=1, color='#A6D785')

# Highlighting the 2007-2008 period
ax.plot(financial_crisis_7_8['Time'],
        financial_crisis_7_8['rolling_mean'],
        linewidth=3, color='#e23d28')

# the following is the result of a process of trial and error, to find the best coordonates. 
# Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 732260
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=11)
    x += 365
y = 1.19   
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(732175, y, rate, alpha=0.5, fontsize=11)
    y += 0.1
    
### Adding a title and a subtitle    
ax.text(732175, 1.67, "Euro-USD rate peaked at 1.59 in 2008 before falling to 1.27 ",
        weight='bold')
ax.text(732175, 1.63, 'Euro-USD exchange rates between 2006 and 2010',
        size=12)

### Adding a signature
ax.text(732175, 1.07, '' + ' '*117 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

### Add some transparency to the grid
ax.grid(alpha=0.5)

plt.show()





```
